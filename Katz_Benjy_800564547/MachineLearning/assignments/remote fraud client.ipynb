```python
import requests
import json
# URL of the API endpoint
url = "https://googlefraudserver-flks2qc3qq-ue.a.run.app/predict"

# JSON body data
data = {
    "TransactionAmt": "999999999",
    "ProductCD": "W",
    "card1": "200",
    "card2": "300",
    "card3": "400",
    "card4": "visa",
    "card5": "400",
    "card6": "credit",
    "P_emaildomain": "gmail.com",
    "email": "5"
}

# Convert the data to JSON format
json_data = json.dumps(data)

# Set the headers (specifying that you're sending JSON)
headers = {'Content-Type': 'application/json'}

# Make the HTTP POST request
response = requests.post(url, data=json_data, headers=headers)

# Print the response
print(response.status_code)
print(response.text)
```
